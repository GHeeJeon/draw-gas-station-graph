```python
import pandas as pd

tmp_list = []
station_files = glob('지역*.xls')
station_files

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_list.append(tmp)
    
stations = pd.concat(tmp_list)
stations.shape
```

```python
stations.info
```

```python
stations.head()
```

```python
stations_sel_gasoline = stations.loc[:, ['상호', '주소', '상표', '셀프여부', '휘발유']]

oil_list = []

for oil in stations['휘발유']:
    oil_list.append('휘발유')

stations_sel_gasoline['종류'] = oil_list

stations_sel_gasoline.reset_index(inplace=True)
stations_sel_gasoline
```

```python
stations_sel_disel = stations.loc[:, ['상호', '주소', '상표', '셀프여부', '경유']]

oil_list = []

for oil in stations['경유']:
    oil_list.append('경유')

stations_sel_disel['종류'] = oil_list

stations_sel_disel.reset_index(inplace=True)
stations_sel_disel
```

```python
stations_sel = pd.concat([stations_sel_gasoline, stations_sel_disel])
stations_sel
```

```python
stations_sel.columns = ['주유소', '주소', '상표', '셀프여부']
stations_sel['종류'].unique()
```

```python
gu_list = []

for address in stations['주소']:
    address_gu = address.split()[1]
    gu_list.append(address_gu)

stations_sel['군/구'] = gu_list
stations_sel
```

```python
stations_sel['군/구'].unique()
```

```python
stations_sel.query("가격 == '-'")
stations_sel
```

```python
stations_sel.query("가격 != '-'")
stations_sel
```

```python
price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)

stations_sel['가격'] = price_list
stations_sel
```

```python
stations_sel = stations_sel.query("가격 != '-'").copy()
stations_sel
```

```python
stations_sel.info()
```

```python
stations_sel.reset_index(inplace=True)
stations_sel
```

```python
del stations_sel['index']
stations_sel
```
